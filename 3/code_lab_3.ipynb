```python
filename = input('Введите название текстового файла: ')
try:
    with open(filename + '.txt', 'r', encoding='utf-8') as file:
        lines = file.readlines()
        n = 0
        for i in lines:
            n += 1
            print(f'Строка {n}: ', i)
except FileNotFoundError:
    print('Ошибка: такого файла не существует')
```

```python
text = input('Введите свой текст: ')
with open('user_input.txt', 'w', encoding='utf-8') as file:
    file.write(text)
    file.close()
```

```python
text = input('Введите текст, который вы хотите добавить: ')
with open('user_input.txt', 'a', encoding='utf-8') as file:
    file.write('\n' + text)
    file.close()
```
