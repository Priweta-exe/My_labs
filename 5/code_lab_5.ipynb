```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
    def get_info(self):
        print(f"Название книги: '{self.title}', Автор: {self.author}, Год издания: {self.year}")
    def set_title(self, new_title):
        self.title = new_title
vim = Book('Война и мир', 'Лев Толстой', 1869)
vim.get_info()
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def get_radius(self):
        return self.radius
    def set_radius(self, new_radius):
        self.radius = new_radius
ring = Circle(10)
print(ring.get_radius())
ring.set_radius(50)
print(ring.get_radius())
```
