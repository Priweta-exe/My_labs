```python
class UserAccount():
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.__password = password
    def set_password(self, new_password):
        self.__password = new_password
    def check_password(self, password):
        return self.__password == password
    def get_password(self):
        return self.__password

Mark = UserAccount('mrmark', 'mrmark@gmail.com', 'secret')
print(Mark.get_password())
Mark.set_password('password')
print(Mark.get_password())
print(Mark.check_password('secret'), Mark.check_password('password'))

#print(Mark.__password)    
```

```python
class Vehicle():
    def __init__(self, make, model, name='Техника'):
        self.make = make
        self.model = model
        self.name = name
    def get_info(self):
        return f'{self.name} марки {self.make} модели {self.model}'

class Car(Vehicle):
    def __init__(self, make, model, fuel_type, name='Машина'):
        super().__init__(make, model)
        self.fuel_type = fuel_type
        self.name = name
    def get_info(self):
        return super().get_info() + f' использует топливо {self.fuel_type}'

uaz = Vehicle('УАЗ', 'Патриот')
print(uaz.get_info())

duster = Car('Renault', 'Duster', 'дизель')
print(duster.get_info())
```
