```python
def greet(name):
    print(f'Здравствуй, {name}')
greet('Ваня')
greet('Никто')
greet('МТУСИ')
```

```python
def square(x):
    return x**2
for i in [2, 4, 3.5, -21]:
    print(i, square(i))
```

```python
def max_of_two(a, b):
    if a >= b:
        return a
    elif a < b:
        return b
for i in [[1, 2], [5, -10], [2.3, -6.8]]:
    print(max_of_two(i[0], i[1]))
```

```python
def describe_person(name, age=30):
    print(f'Перед нами человек по имени {name} и возрастом {age} лет')
describe_person('Алина', 15)
describe_person('Артем', 25)
describe_person('Витя')
```

```python
from math import ceil
def is_prime(n):
    for i in range(2, ceil(n**0.5)+1):
        if n % i == 0:
            return False
    return True
for i in [4, 6, 11, 153, 63, 61, 80, 1001]:
    print(i, is_prime(i))
```
